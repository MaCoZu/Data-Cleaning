```python
import matplotlib.pyplot as plt

# Filter data for specific countries
china = output[output.Country == 'China']
germany = output[output.Country == 'Germany']

plt.style.use('ggplot')

# Create a single figure
plt.figure(figsize=(10, 6))

# Plot China's data
plt.plot(china['Year'], china['ValueUSD'], label='China')

# Plot Germany's data
plt.plot(germany['Year'], germany['ValueUSD'], label='Germany')

# Add labels, legend, and title
plt.xlabel('Year')
plt.ylabel('Value (USD)')
plt.title('Value (USD) Over Years for China and Germany')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
